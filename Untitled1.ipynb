```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_df = pd.DataFrame(pd.read_csv('train.csv'))
test_df = pd.DataFrame(pd.read_csv('test.csv'))
```

```python
train_df.head(15)
```

```python
train_df.isnull().sum()
```

```python
train_df = train_df.drop(["Cabin"], axis = 1)
```

```python
train_df.head(15)
```

```python
train_df['Age'].fillna(train_df['Age'].median(), inplace=True)
```

```python
train_df.head(15)
```

```python
train_df['Embarked'].fillna(train_df['Embarked'].mode(), inplace = True)
```

```python
train_df.info()
```

```python
train_df = train_df.drop(["PassengerId", "Name" , "Ticket" , "Fare"], axis = 1)
```

```python
train_df.info()
```

```python
train_df.head()
```

# Chuyển dạng String sang dạng Number

```python
from sklearn.preprocessing import LabelEncoder
```

```python
col_str = train_df.drop(train_df.select_dtypes(exclude=['object']), axis=1).columns
print(col_str)
enc1 = LabelEncoder()
train_df[col_str[0]] = enc1.fit_transform(train_df[col_str[0]].astype("str"))
enc2 = LabelEncoder()
train_df[col_str[1]] = enc2.fit_transform(train_df[col_str[1]].astype("str"))
```

```python
print(train_df[col_str[0]])
```

```python
train_df.info()
```

```python
train_df.head()
```

# Training

```python
X = train_df.drop(['Survived'], axis=1)
Y = train_df['Survived']
X.info()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()  
lr = model.fit(X_train, Y_train)
```

```python
lr
```

# TEST

```python
Y_pred = model.predict(X_test)

pred_df = pd.DataFrame({'Thực tế': Y_test, 'Dự đoán': Y_pred})
pred_df.head(15)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(Y_pred, Y_test))
```
